```python
#Bob will end the game immediately, so only Alices first move matters.

#Swaps of same parity: only increase f(a) by cost (r - l).

#Swaps of different parity: increase f(a) by cost plus 2 * value difference.

#Prefix minimums are used to efficiently find the optimal swap across the array.

#This gives maximum f(a) for Alice assuming optimal play.
```

```python
tt = int(input())

for _ in range(tt):
    n = int(input())
    a = list(map(int, input().split()))

    ans = 0
    for i in range(n):
        if i % 2:
            ans -= a[i]
        else:
            ans += a[i]

    init = ans

    for i in range(n):
        ans = max(ans, init + i - (i % 2))

    min_even = float('inf')
    min_odd = float('inf')

    for i in range(n):
        if i % 2:
            ans = max(ans, init + i + a[i] + a[i] - min_even)
            min_odd = min(min_odd, i - a[i] - a[i])
        else:
            ans = max(ans, init + i - a[i] - a[i] - min_odd)
            min_even = min(min_even, i + a[i] + a[i])

    print(ans)
```
