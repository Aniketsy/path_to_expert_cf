```python
T = int(input())
for _ in range(T):
    k, x, a = map(int, input().split())

    req = 1
    k -= 1
    possible = True

    for _ in range(x + 1):
        cur = (req + k - 1) // k  # minimum bet to cover previous losses
        a -= cur                   # deduct from current coins
        req += cur                 # update required coins for next round
        if a < 0:                  # cannot survive
            possible = False
            break

    print("YES" if possible else "NO")
```
