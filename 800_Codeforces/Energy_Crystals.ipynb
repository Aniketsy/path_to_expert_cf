```python
t = int(input())
for _ in range(t):
    x = int(input())
    crystals = [0, 0, 0]
    steps = 0
    while min(crystals) < x:
      crystals.sort()
      m = crystals[1]
      crystals[0] = min(x,2*m+1)
      steps +=1
    print(steps)
```
